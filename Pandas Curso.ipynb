**Filtrado con |, es decir, una u otra.**
Se pueden combinar tanto como se requiera

```python
citi[(citi['credit_amount'] < 300) | (citi['credit_amount'] > 15000)]
```

>Note: When combining multiple conditional statements, each condition must be surrounded by parentheses (). Moreover,
you can not use or/and but need to use the or operator | and the and operator &.

```python
citi[(citi['credit_amount'] < 300) | (citi['target'] == 'bad')]
```

```python
citi[(citi['credit_amount'] < 900) & (citi['target'] == 'bad')]
```

**Filas y Columnas espeficias**

```python
citi[citi['credit_amount'] > 15000]
```

```python
citi.loc[citi['credit_amount'] > 15000,"foreign_worker"]
```

>En este caso, se crea un subconjunto de filas y columnas de una sola vez y ya no es suficiente usar corchetes de selecci√≥n []. Los operadores loc/iloc son obligatorios delante de los corchetes de selecci√≥n []. Al usar loc/iloc, la parte anterior a la coma son las filas que desea y la parte posterior a la coma son las columnas que desea seleccionar.

Estoy interesado en las filas 10 a 15 y las columnas 3 a 5.

```python
citi.iloc[10:16,2:5] #empiza en cero
```
